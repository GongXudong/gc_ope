```python
import numpy as np
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt

# 1. 创建两个明显不同的数据集
# 数据集A: 均值为0的正态分布
data_A = np.random.randn(100, 1) 
# 数据集B: 均值为10的正态分布
data_B = np.random.randn(100, 1) + 10

# 2. 初始化一个KernelDensity对象
kde = KernelDensity(kernel='gaussian', bandwidth=0.5)

# --- 第一次拟合 ---
print("--- 第一次拟合 (使用 data_A) ---")
kde.fit(data_A)

# 在一个靠近 data_A 均值的点 (0.0) 评估密度
score_A_point = kde.score_samples([[0.0]])
print(f"在点 0.0 处的对数概率密度 (基于 data_A): {score_A_point[0]:.2f}")
# 这个值应该相对较高（接近0，因为对数概率）

# --- 第二次拟合 ---
print("\n--- 第二次拟合 (使用 data_B) ---")
kde.fit(data_B)

# 再次在同一个点 (0.0) 评估密度
score_A_point_after_B = kde.score_samples([[0.0]])
print(f"在点 0.0 处的对数概率密度 (基于 data_B): {score_A_point_after_B[0]:.2f}")
# 这个值应该会变得非常低（一个很大的负数），因为模型已经“忘记”了 data_A

# 在一个靠近 data_B 均值的点 (10.0) 评估密度
score_B_point = kde.score_samples([[10.0]])
print(f"在点 10.0 处的对数概率密度 (基于 data_B): {score_B_point[0]:.2f}")
# 这个值应该相对较高

kde.fit(data_A)
score_A_point = kde.score_samples([[0.0]])
print(f"在点 0.0 处的对数概率密度 (基于 data_A): {score_A_point[0]:.2f}")

kde.fit(data_B)
score_A_point_after_B = kde.score_samples([[0.0]])
print(f"在点 0.0 处的对数概率密度 (基于 data_B): {score_A_point_after_B[0]:.2f}")
# 这个值应该会变得非常低（一个很大的负数），因为模型已经“忘记”了 data_A

# 在一个靠近 data_B 均值的点 (10.0) 评估密度
score_B_point = kde.score_samples([[10.0]])
print(f"在点 10.0 处的对数概率密度 (基于 data_B): {score_B_point[0]:.2f}")
```
